```python
import numpy as np
import pandas as pd
```

```python
df_ventas = pd.read_csv("Dataset-Ventas.csv", encoding="UTF-8")

df_ventas["ID Pedido"] = df_ventas["ID Pedido"].astype(np.int64)
df_ventas.rename(columns={"ID Pedido": "ID_Pedido"}, inplace=True)
df_ventas.info()

df_ventas
```

```python
import pymysql

conn = pymysql.connect( user = "admin",
                       password = "abc123",
                       host = "172.18.0.2",
                       db = "bbdd")

query = "select * from DIRECCIONES;"

df_direcciones = pd.read_sql(query, con = conn)

conn.close()
df_direcciones.info()

```

```python
df_ventas["ID_Pedido"] = df_ventas["ID_Pedido"].astype(str)
df_direcciones["ID_pedido"] = df_direcciones["ID_pedido"].astype(str)

# merge para unir ambas fuentes de datos por la columna ID_Pedido
df_completo = pd.merge(df_ventas, df_direcciones, left_on="ID_Pedido", right_on="ID_pedido", how="inner") # Una columna es ID_Pedido otra ID_pedido

df_completo
```

Dado que hay multiples registros con el mismo ID_Pedido, y vamos a utilizar ID Factura como índice tal como se indica en el enunciado, eliminamos las columnas `ID_Pedido`

```python
df_completo = df_completo.drop(columns = ['ID_Pedido', 'ID_pedido'])
df_completo.set_index("ID Factura", inplace = True)
df_completo.head()
```

También se prescindirá de las columnas `Comercio` y `Fecha` pues son otros de los requisitos del enunciado

```python
df_completo = df_completo.drop(columns = ['Comercio', 'Fecha'])
df_completo.head()
```

Ahora que tenemos el dataframe con las columnas que queremos, haremos la comprobación de `valores nulos` y `duplicados`.

```python
# comprobación de nulos
df_completo.isnull().sum()
```

Para las columnas `Tipo cliente` y `Sexo` se sustuirán los valores nulos por el valor "Desconocido".

```python
print(f"{df_completo['Tipo cliente'].unique()}") # valores que puede tomar 'Tipo Cliente'
```

```python
df_completo['Tipo cliente'] = df_completo['Tipo cliente'].fillna("Desconocido")
print(f"{df_completo['Tipo cliente'].unique()}") 
```

```python
df_completo['Sexo'] = df_completo['Sexo'].fillna("Desconocido")
print(f"{df_completo['Sexo'].unique()}")
```

La columna `Precio Unitario` presenta un único valor nulo, que será sustituido por la media de esa columna para afectar lo mínimo posible al resultado final. Sin embargo, los datos de esta columna son de tipo `object` y deben ser convertidos primeramente a `float` para poder obtener la media y operar con ellos.

```python
# Convierte la columna 'Precio Unitario' de object a float, reemplazando valores no numéricos por NaN
df_completo['Precio Unitario'] = pd.to_numeric(df_completo['Precio Unitario'], errors='coerce')

# Rellena los valores NaN con la media de la columna
df_completo['Precio Unitario'] = df_completo['Precio Unitario'].fillna(df_completo['Precio Unitario'].mean())
```
