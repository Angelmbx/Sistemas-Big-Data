```python
import requests
import pandas as pd  # Para trabajar con tablas de datos (opcional)

# Tu API Key de NewsAPI
API_KEY = "TU_API_KEY"
URL = "https://newsapi.org/v2/top-headlines"

# Parámetros para la solicitud
params = {
    "country": "es",  # Puedes cambiar el país us, ext
    "category": "tecnología",  # Cambia la categoría si quieres (ejemplo: 'negocio', 'deportes')
    "apiKey": API_KEY        #se obtiene aquí      https://newsapi.org/ 
}

# Función para obtener los titulares de las noticias

def obtener_titulares():
    try:
        # Realizar la solicitud a la API
        respuesta = requests.get(URL, params=params)
        respuesta.raise_for_status()  # Verifica si hubo algún error en la solicitud
        
        #devuelve los datos
        datos = respuesta.json()
        
        if datos["status"] == "ok":
            # Crear una lista vacía para almacenar las noticias
            noticias = []
            
            #datos = {
                    #"articles": [
                    #        {"title": "Noticia 1", "source": {"name": "Fuente 1"}, "url": "http://noticia1.com"},
                    #        {"title": "Noticia 2", "source": {"name": "Fuente 2"}, "url": "http://noticia2.com"}
                    #            ]
                    #}
            
            
            # Iterar por cada artículo en los resultados
            for articulo in datos["articles"]:
                noticia = {
                    "Título": articulo["title"],
                    "Fuente": articulo["source"]["name"],
                    "URL": articulo["url"]
                }
                noticias.append(noticia)  # Agregar la noticia a la lista
            
            # Mostrar como tabla usando pandas
            df_noticias = pd.DataFrame(noticias)
            return df_noticias
        else:
            print("Error al obtener titulares:", datos["message"])
    except Exception as e:
        print("Error:", e)

# Ejecuta esta celda para obtener y visualizar titulares
df_titulares = obtener_titulares()
if df_titulares is not None:
    display(df_titulares)  # Muestra la tabla en JupyterLab
```
